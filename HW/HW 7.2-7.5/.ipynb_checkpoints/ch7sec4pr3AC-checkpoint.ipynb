```python
import pandas as pd
from string import ascii_lowercase
from numpy import zeros, diag, diagflat, dot


# Example Systems in matrix and vector form 
a = [[3, -1, 1],
    [3, 6, 2],
    [3, 3, 7]]
a1 = [1,0,4]

b = [[10, -1, 0],
    [-1, 10, -2],
    [0, -2, 10]]
b1 = [9, 7, 6]

c = [[10, 5, 0, 0],
    [5, 10, -4, 0],
    [0, -4, 8, -1],
    [0, 0, -1, 5]]
c1 = [6,25,-11,11]

d = [[4, 1, -1, 1],
    [1, 4, -1, 1],
    [-1, -1, 5, 4],
    [1, -1, 1, 3]]
d1 = [2,-1,0,1]

e = [[4, 1, 1, 0, 1],
    [-1, -3, 1, 1, 0],
    [2, 1, 5, -1, -1],
    [-1, -1, -1, 4, 0],
    [0, 2, -1, 1, 4]]
e1 = [6,6,6,6,6]

f = [[4, -1, 0, 0, 0, 0],
    [-1, 4, -1, 0, 0, 0],
    [0, -1, 4, 0, 0, 0],
    [0, 0, 0, 4, -1, 0], 
    [0, 0, 0, -1, 4, -1],
    [0, 0, 0, 0, -1, 4 ]]
f1 = [0,5,0,6,-2, 6]

def Matrix_norm(A):
    n, l = len(A), []          
    for i in range(n):
        c = 0                   
        for j in range(n):
            c += abs(A[i][j])   
        l.append(c)            
    return max(l) 

def jacobi(A,b,N=25,x=None, tol = 1e-3):
    """Solves the equation Ax=b via the Jacobi iterative method."""
    # Create an initial guess if needed                                                                                                                                                            
    if x is None:
        x = zeros(len(A[0]))

    # Create a vector of the diagonal elements of A                                                                                                                                                
    # and subtract them from A                                                                                                                                                                     
    D = diag(A)
    R = A - diagflat(D)

    # Iterate for N times                                                                                                                                                                          
    for i in range(N):
        x = (b - dot(R,x)) / D
    return x

def jacobi_sol(part, mat, b):
    print('Part {} ------------------------------------------------\nA = \n{}\n\n b = \n{} \n\n Solution from Jacobi: {}\n\n\n '.format(part, pd.DataFrame(mat),pd.DataFrame(b), jacobi(mat, b)))
    
matrices = [a,b,c,d,e,f]
mat_b = [a1,b1,c1,d1,e1,f1]
for i, c, b in zip(matrices, ascii_lowercase, mat_b):
    jacobi_sol(c, i, b)
```
