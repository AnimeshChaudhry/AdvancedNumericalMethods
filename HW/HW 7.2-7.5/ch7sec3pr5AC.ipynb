```python
from itertools import chain
import math
import pandas as pd
from string import ascii_lowercase



# Example Matrices from the book
a = [[2, -1],
    [-1, 2]]

b = [[0, 1],
    [1, 1]]

c = [[0, 1/2],
    [1/2, 0]]

d = [[1, 1],
    [2, 2]]

e = [[2, 1, 0],
    [1, 2, 0],
    [0, 0, 3]]

f = [[-1, 2, 0],
    [0, 3, 4],
    [0, 0, 7]]

g = [[2, 1, 1],
    [2, 3, 2],
    [1, 1, 2]]

h = [[3, 2, -1],
    [1, -2, 3],
    [2, 0, 4]]




def Matrix_norm_two(mat): 
    mat = list(chain.from_iterable(zip(*mat)))
    s = 0
    for i in mat:
        s+=i**2
    return math.sqrt(s)

def print_norm(part, mat):
    print('Part {} ------------------------------------------------\n{}\n The L2 norm is: {}\n'.format(part, pd.DataFrame(mat), Matrix_norm_two(mat)))
    
matrices = [a,b,c,d,e,f,g,h]
for i, c in zip(matrices, ascii_lowercase):
    print_norm(c, i)

```
