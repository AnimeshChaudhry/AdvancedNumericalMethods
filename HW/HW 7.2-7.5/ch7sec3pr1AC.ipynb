```python
from numpy import linalg as LA
import pandas as pd
from string import ascii_lowercase


# Example Matrices from the book
a = [[2, -1],
    [-1, 2]]

b = [[0, 1],
    [1, 1]]

c = [[0, 1/2],
    [1/2, 0]]

d = [[1, 1],
    [2, 2]]

e = [[2, 1, 0],
    [1, 2, 0],
    [0, 0, 3]]

f = [[-1, 2, 0],
    [0, 3, 4],
    [0, 0, 7]]

g = [[2, 1, 1],
    [2, 3, 2],
    [1, 1, 2]]

h = [[3, 2, -1],
    [1, -2, 3],
    [2, 0, 4]]





def eigen_print(part, mat):
    print('Part {} ------------------------------------------------\n{}\n The eiven value(s) is(are): {}\n The Eigenvector(s) is(are): {}\n'.format(part, pd.DataFrame(mat), LA.eig(mat)[0], LA.eig(mat)[1]))
    

matrices = [a,b,c,d,e,f,g,h]
for i, c in zip(matrices, ascii_lowercase):
    eigen_print(c, i)
```
