```python
'''
(2)

Find the eigenvalues and associated eigenvectors for the following 3 Ã— 3 matrices. Is there a set of
three linearly independent eigenvectors?
'''
import numpy as np
from numpy import linalg as LA
from string import ascii_lowercase

a = np.array([[2, -3, 6], [0, 3, -4], [0, 2, -3]])
b = np.array([[1, 0, 0], [-1, 0, 1], [-1, -1, 2]])
c = np.array([[2, 0, -1], [0, 2, 0], [1, 0, 2]])
d = np.array([[2, -1, -1], [-1, 2, -1], [-1, -1, 2]])
e = np.array([[1, 1, 1], [1, 1, 0], [1, 0, 1]])
f = np.array([[2, 1, 1], [1, 2, 1], [1, 1, 2]])

def print_egn(A):
    print(f"{A} \nEigenValue(s): ----> {LA.eig(A)[0]}\nEigenvectors: ----> \n{LA.eig(A)[1]}\nDeterminent of eigenvector matrix: {np.linalg.det(np.array(LA.eig(A)[1]))}\n\n")

matrices = [a,b,c,d, e,f]
for i in matrices:
    print_egn(i)
print(f"If the determinent of eigenvector matrix is a number fairly close to zero, then the vectors are not L.I., otherwise, they are")
```
